```python
import pandas as pd
import numpy as np
from pathlib import Path
Prof_loss = []
total = 0
average_chg = 0
g_increase = 0
g_decrease = 0
data_f = []
## This is a test
```

```python
f = Path("budget_data.csv")
```

```python
data_f = pd.read_csv(f)
```

```python
data_f
```

```python
total = data_f['Date'].count()
```

```python
Prof_loss = data_f['Profit/Losses'].sum()   ## reads the Prof/loss column into memory and assigns the value 
```

```python
data_f['Change'] = data_f['Profit/Losses'].diff()
```

```python
average_chg = round(data_f['Change'].mean(),2)
```

```python
max_incr= data_f['Change'].max()
```

```python
incr_date = data_f.loc[data_f["Change"] == max_incr]
```

```python
min_incr = data_f['Change'].min()
decr_date = data_f.loc[data_f["Change"] == min_incr ]
```

```python
print("Financial Analysis\n", "------------------")
print("Total Months: ", total)
print("Total profit/losses: $ ", Prof_loss)   
print("Average Change over the period was: $",average_chg)
print("The Greatest increase in Profits occurred in ",incr_date ['Date'].values[0]," with a value of  $",max_incr)
print("The Greatest decrease in Profits occurred in ",decr_date ['Date'].values[0]," with a value of  $",min_incr)
```

```python
report = f"""
Financial Analysis
------------------
Total Months: {total}
Total profit/losses: ${Prof_loss}
Average Change over the period was: ${average_chg}
The Greatest increase in Profits occurred in {incr_date ['Date'].values[0]} with a value of  ${max_incr}
The Greatest decrease in Profits occurred in {decr_date ['Date'].values[0]} with a value of  ${min_incr}
"""
print(report)
```

```python
with open('output.txt', 'w') as f:
    f.write(report)
    
```
